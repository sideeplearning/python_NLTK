# MT_methodã§ç•°å¸¸æ¤œçŸ¥ã€ãƒ‡ãƒ¼ã‚¿åˆ†é¡ž

```python
import numpy as np
import matplotlib.pyplot as plt

divide = 2 #çŠ¶æ…‹é‡ã®æ•°
R = np.zeros((divide,divide))    #ç›¸é–¢è¡Œåˆ—
invR = np.zeros((divide,divide)) #ç›¸é–¢è¡Œåˆ—ã®é€†è¡Œåˆ—
avg = np.zeros(divide)           #å¹³å‡å€¤
std = np.zeros(divide)           #æ¨™æº–åå·®
make = 0

p = 0.95        #ãƒžãƒãƒ©ãƒŽãƒ“ã‚¹è·é›¢p=0.95ã§2Ïƒ
md_sikii = 2.448#MDã®é–¾å€¤95%ã§2.448
div = 50        #Mtæ¥•å††ã®åˆ†å‰²æ•°

def maha(x):
    global make, R, invR, avg, std

    N, _ = x.shape #Nã¯ãƒ‡ãƒ¼ã‚¿æ•°
    xx = np.copy(x)
    xx = np.array(xx,dtype="float32")
    x_return = []

    #å„çŠ¶æ…‹é‡ã‹ã‚‰å¹³å‡å€¤ã‚’å¼•ã
    for i in range(divide):
        if make == 0:
            avg[i] = np.mean(x[:,i])
        for j in range(N):
            xx[j,i] = xx[j,i] - avg[i]

    #å„çŠ¶æ…‹é‡ã‚’æ¨™æº–åå·®ã§å‰²ã‚‹
    for i in range(divide):
        if make == 0:
            std[i] = np.std(x[:,i])
        for j in range(N):
            xx[j,i] = xx[j,i] / std[i]

    #make=0ã®ã¨ãã ã‘è¨ˆç®—
    if make == 0:
        R = np.corrcoef(xx.transpose())
        invR = np.linalg.inv(R)
        make = 1

    #MD^2ã®è¨ˆç®—
    for i in range(N):
        d0 = xx[i,:]
        d1 = np.dot(d0,invR)
        d2 = np.dot(d1,d0)/divide
        x_return.append(d2)

    return x_return

```

```python
data_ = data_list[2].drop('label', axis =1)
data_
```

```python
#PCA
label=data_list[2].label
print(label.shape, data_list[2].shape)
data_list[2] =data_list[2].drop('label', axis =1)
data_list[2]

X_train,x_test,Y_train,y_test=train_test_split(data_list[2], label , test_size = 0.3, random_state = 666)
#data_list[2]np=data_list[2].values
#data_list[2]np.shape
#data_list[2]np
pca=PCA(n_components=3)
pca.fit(X_train)
pca_X=pca.transform(X_train)

print("data_list[2]_shape:"+ str(data_list[2].shape))
print("X_train:"+ str(X_train.shape))
print("Y_train:"+ str(Y_train.shape))

print("pca.singular_values_:"+ str(pca.singular_values_))
print("pca.explained_variance_ratio_:"+ str(pca.explained_variance_ratio_))
print("pca_MEAN:"+ str(pca.mean_))
print("pca_X:"+ str(pca_X))

ev_ratio = pca.explained_variance_ratio_
ev_ratio
ev_ratio = np.hstack([0,ev_ratio])
plt.title("ev_ratio")
plt.plot(ev_ratio)
plt.grid(True)
plt.xlabel('n_components')
plt.ylabel('ev_ratio')
plt.show()
#plt.figure(figsize=(30,30))

ev_ratio_sum = np.hstack([0,ev_ratio.cumsum()])
plt.title("ev_ratio_sum")
plt.plot(ev_ratio_sum)
plt.grid(True)
plt.xlabel('n_components')
plt.ylabel('ev_ratio_sum')
plt.show()
#plt.figure(figsize=(30,30))

ev_ratio
```
