# MT_methodで異常検知、データ分類

```python
import numpy as np
import matplotlib.pyplot as plt

divide = 2 #状態量の数
R = np.zeros((divide,divide))    #相関行列
invR = np.zeros((divide,divide)) #相関行列の逆行列
avg = np.zeros(divide)           #平均値
std = np.zeros(divide)           #標準偏差
make = 0

p = 0.95        #マハラノビス距離p=0.95で2σ
md_sikii = 2.448#MDの閾値95%で2.448
div = 50        #Mt楕円の分割数

def maha(x):
    global make, R, invR, avg, std

    N, _ = x.shape #Nはデータ数
    xx = np.copy(x)
    xx = np.array(xx,dtype="float32")
    x_return = []

    #各状態量から平均値を引く
    for i in range(divide):
        if make == 0:
            avg[i] = np.mean(x[:,i])
        for j in range(N):
            xx[j,i] = xx[j,i] - avg[i]

    #各状態量を標準偏差で割る
    for i in range(divide):
        if make == 0:
            std[i] = np.std(x[:,i])
        for j in range(N):
            xx[j,i] = xx[j,i] / std[i]

    #make=0のときだけ計算
    if make == 0:
        R = np.corrcoef(xx.transpose())
        invR = np.linalg.inv(R)
        make = 1

    #MD^2の計算
    for i in range(N):
        d0 = xx[i,:]
        d1 = np.dot(d0,invR)
        d2 = np.dot(d1,d0)/divide
        x_return.append(d2)

    return x_return

```

```python
import numpy as np
import matplotlib.pyplot as plt

divide = 2 #状態量の数
R = np.zeros((divide,divide))    #相関行列
invR = np.zeros((divide,divide)) #相関行列の逆行列
avg = np.zeros(divide)           #平均値
std = np.zeros(divide)           #標準偏差
make = 0

p = 0.95        #マハラノビス距離p=0.95で2σ
md_sikii = 2.448#MDの閾値95%で2.448
div = 50        #Mt楕円の分割数

def maha(x):
    global make, R, invR, avg, std

    N, _ = x.shape #Nはデータ数
    xx = np.copy(x)
    xx = np.array(xx,dtype="float32")
    x_return = []

    #各状態量から平均値を引く
    for i in range(divide):
        if make == 0:
            avg[i] = np.mean(x[:,i])
        for j in range(N):
            xx[j,i] = xx[j,i] - avg[i]

    #各状態量を標準偏差で割る
    for i in range(divide):
        if make == 0:
            std[i] = np.std(x[:,i])
        for j in range(N):
            xx[j,i] = xx[j,i] / std[i]

    #make=0のときだけ計算
    if make == 0:
        R = np.corrcoef(xx.transpose())
        invR = np.linalg.inv(R)
        make = 1

    #MD^2の計算
    for i in range(N):
        d0 = xx[i,:]
        d1 = np.dot(d0,invR)
        d2 = np.dot(d1,d0)/divide
        x_return.append(d2)

    return x_return

```

```python
#楕円のデータ
low_2 = np.corrcoef(data_2[:,0],data_2[:,1])[0,1]
low_10 = np.corrcoef(data_10[:,0],data_10[:,1])[0,1]
```

```python
#楕円半径の計算　Poe

curve_c_2 = np.zeros((2,div+1))
for i in range(div+1):
    r = (-2*(1-low_2**2)*np.log(1-p)/(1-2*low_2*np.sin(i*2*np.pi/div)*np.cos(i*2*np.pi/div)))**0.5
    curve_c_2[0,i] = avg_2_x + std_2_x *r*np.cos(i*2*np.pi/div)
    curve_c_2[1,i] = avg_2_y + std_2_y *r*np.sin(i*2*np.pi/div)
    
    print(r)
curve_c_2
```

```python
#楕円半径の計算　Twain

curve_c_10 = np.zeros((2,div+1))
for i in range(div+1):
    r = (-2*(1-low_10**2)*np.log(1-p)/(1-2*low_10*np.sin(i*2*np.pi/div)*np.cos(i*2*np.pi/div)))**0.5
    curve_c_10[0,i] = avg_10_x + std_10_x*r*np.cos(i*2*np.pi/div)
    curve_c_10[1,i] = avg_10_y + std_10_y*r*np.sin(i*2*np.pi/div)
    print(r)
curve_c_10
```
